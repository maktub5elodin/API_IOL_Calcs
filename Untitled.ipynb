```python
def GetInstruments(panel=DATOS_Paneles[0]):
    
    InstrumentsResponse = r.get(url=URL_IOL_All_Instruments, params={ "cotizacionInstrumentoModel.instrumento" : panel, "cotizacionInstrumentoModel.pais" : "argentina"}, headers={"Authorization" : "Bearer " + TokenBearer}).json()["titulos"]
    
    return [i["simbolo"] for i in InstrumentsResponse]
```

```python
Instruments0 = GetInstruments()
Instruments1 = GetInstruments(DATOS_Paneles[1])
Instruments2 = GetInstruments(DATOS_Paneles[2])
#Instruments3 = GetInstruments(DATOS_Paneles[3])
```

```python
def Cotizacion(simbolo, mercado = "BCBA", plazo = "t2"):
    
    CotizacionParametros = {"mercado"       : mercado,
                            "simbolo"       : simbolo, 
                            "model.simbolo" : simbolo, 
                            "model.mercado" : mercado, 
                            "model.plazo"   : plazo}
    
    return r.get(url=URL_IOLCotizacion, params=CotizacionParametros, headers={"Authorization" : "Bearer " + TokenBearer}).json()
```

```python
def GetDFvolume(inst=Instruments0, tim = "t2"):
    
    DFvolume = pd.DataFrame(index=inst, columns=["montoOperado", "percentOperado"])
    
    for i in inst:
        DFvolume["montoOperado"][i] = round(Cotizacion(i, plazo = tim)["montoOperado"] / 1000000, 2)

    for i in inst:
        DFvolume["percentOperado"][i] = round(DFvolume["montoOperado"][i] / DFvolume.sum()["montoOperado"] * 100, 2)
        
    DFvolume.sort_values("percentOperado", ascending=False, inplace=True)
        
    return DFvolume
```

```python
def tcventa(ars, usd, fee="platinum"):
    return round((ars * (1 - DATOS_Comisiones[fee]))/(usd * (1 + DATOS_Comisiones[fee])), 2)

def tccompra(ars, usd, fee="platinum"):
    return round((ars * (1 + DATOS_Comisiones[fee]))/(usd * (1 - DATOS_Comisiones[fee])), 2)

# Vale la pena crear variables locales para cada parte del numerador y denominador?
# revisar los decimales
#FALTA INCORPORAR LOS DERECHOS DE MERCADO
#FALTA INCLUIR VALORES MINIMOS
# Compra no tiene en cuenta el parking
```

```python
ListMEP = [i for i in Instruments2 if i.upper().endswith("D")]
ListMEPstrip = [i.rstrip(i[-1]) for i in ListMEP]

ListCCL = [i for i in Instruments2 if i.upper().endswith("C")]
ListCCLstrip = [i.rstrip(i[-1]) for i in ListCCL]

ListExclude = []

for i in ListMEPstrip:
    if i not in Instruments2:
        ListExclude.append(i + "D")
        
for i in ListCCLstrip:
    if i not in Instruments2:
        ListExclude.append(i + "C")

for i in ListExclude: 
    if i in ListMEP:
        ListMEP.remove(i)

    if i in ListCCL:
        ListCCL.remove(i)

ListMEPstrip = [i.rstrip(i[-1]) for i in ListMEP]
ListCCLstrip = [i.rstrip(i[-1]) for i in ListCCL]

List_Intersect = [i for i in ListMEPstrip if i in ListCCLstrip]
List_Unique = [i for i in ListMEPstrip if i not in ListCCLstrip]
```

```python
def GetDFFX(fee="platinum"):
    
    DFFX = pd.DataFrame(index=List_Intersect, columns=["Compra_"+fee, "Venta_"+fee])
    
    for i in List_Intersect:
        TickerARS = Cotizacion(i, plazo = "t0")
        TickerUSD = Cotizacion(i + "D", plazo = "t0")
        try:
            DFFX["Compra_"+fee][i] = tcventa(TickerARS["puntas"][0]["precioCompra"], TickerUSD["puntas"][0]["precioVenta"], fee)
            
            DFFX["Venta_"+fee][i] = tccompra(TickerARS["puntas"][0]["precioVenta"], TickerUSD["puntas"][0]["precioCompra"], fee)
        except:
          continue
        
    return DFFX        
```

```python
def GetDF_Rendimientos():
    DF_Rendimientos = (pd.DataFrame(index=DATOS_Portfolio_Activos, columns=["Rendimiento%", "Tipo_de_Titulo"]))

    for i in DATOS_Portfolio:

        iticker = i["titulo"]["simbolo"]
        itipo = i["titulo"]["tipo"]
        irendimiento = i["gananciaPorcentaje"]

        DF_Rendimientos["Rendimiento%"][iticker] = irendimiento
        DF_Rendimientos["Tipo_de_Titulo"][iticker] = itipo

    return DF_Rendimientos
```

```python
def GetDF_Weight(Tipo_de_Titulo = DATOS_Portfolio_Paneles[0]):

    DF_Weight = (pd.DataFrame(index=DATOS_Portfolio_Activos, columns=["Valor"]))

    for i in DATOS_Portfolio:

        iticker = i["titulo"]["simbolo"]
        itipo = i["titulo"]["tipo"]
        ivalorizado = i["valorizado"]

        if itipo == Tipo_de_Titulo:

            DF_Weight["Valor"][iticker] = ivalorizado

        else:
          continue 

    DF_Weight.dropna(inplace=True)

    TotalPortfolio = DF_Weight.sum()["Valor"]

    DF_Weight["Peso"] = None

    for i in DF_Weight.index:

      DF_Weight["Peso"][i] = round(DF_Weight["Valor"][i] / TotalPortfolio * 100, 2)

    DF_Weight.sort_values(by=["Valor"], ascending=False, inplace=True)

    return DF_Weight
```

```python
def GetDF_Hist_Vol(simbolo, mercado = "BCBA", fechaDesde = "2022-11-01", fechaHasta = "2022-12-01"):

  URL_IOL_Serie_Historica = f"https://api.invertironline.com/api/v2/{mercado}/Titulos/{simbolo}/Cotizacion/seriehistorica/{fechaDesde}/{fechaHasta}/sinAjustar"
  dato = r.get(url=URL_IOL_Serie_Historica, headers={"Authorization" : "Bearer " + TokenBearer}).json()

  indice = [i["fechaHora"][0:10] for i in dato]

  DF_Hist_Vol = pd.DataFrame(index=indice, columns=[simbolo])

  for i in dato:

    icuanDo = i["fechaHora"][0:10]
    icuanTo = round(i["montoOperado"] / 1000000, 2)

    DF_Hist_Vol[simbolo][icuanDo] = icuanTo

  return DF_Hist_Vol
```
